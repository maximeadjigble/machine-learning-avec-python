```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot([1, 2, 3, 4], [1, 3 , 2, 4])
plt.title("Prix de stock")
plt.xlabel("Temps (s)")
plt.ylabel("Prix ($)")
plt.show()
```

```python
t = [1,2,3,4,5]
prix_microsoft = [1,3,2,4,8]
prix_apple = [2,4,4,3,6]
plt.plot(t, prix_microsoft, label="microsoft")
plt.plot(t, prix_apple, label="apple")
plt.title("Prix de stock")
plt.xlabel("Temps (s)")
plt.ylabel("Prix ($)")
plt.legend()
plt.show()
```
