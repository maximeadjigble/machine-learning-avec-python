```python
X = [1, 2, -6, 6]
W = [1, 10, 1, 2]
b = 2

y = 0
for xi, wi in zip(X, W):
    y = y + xi*wi

y = y + b
y
```

```python
def lin_reg(b, W, X):
    y = 0
    for xi, wi in zip(X, W):
        y = y + xi*wi
    y = y + b
    return y
```

```python
b = 4
W = [5, 2]
X = [1, 1]

y = lin_reg(b, W, X)
y
```

```python
import random
import matplotlib.pyplot as plt
%matplotlib inline

X1 = []
X2 = []
Y = []

for i in range(1000):
    x1 = random.uniform(-5, 5)
    x2 = random.uniform(-5, 5)
    X = [x1, x2]
    y = lin_reg(b, W, X)
    X1.append(x1)
    X2.append(x2)
    Y.append(y)
    
ax = plt.axes(projection="3d")
ax.view_init(30, 100)
ax.scatter3D(X1, X2, Y)
plt.show()
```
